**Διαχωρισμός του scaled dataset σε Train-Test**   
Μετά από σχολιασμό στο μάθημα, εκτελούμε πρώτα διαχωρισμό και μετά MinMaxScaling ώστε να μην έχουμε data leakage.

```python
from sklearn.model_selection import train_test_split
```

```python
X_ = df1.likes
y_ = df1.views 
X_tr, X_te, y_tr, y_te = train_test_split(X_,y_,test_size = 0.2, random_state = 24 )
```

```python
print('X_train shape:',X_tr.shape)
print('X_test shape:',X_te.shape)
print('y_train shape:',y_tr.shape)
print('y_test shape:',y_te.shape)
```

#### Για το Min-Max Scaling, χρησιμοποιούμε την αντίστοιχη συνάρτηση από το Sklearn, που πραγματοποιεί τον εξής μετασχηματισμό: $$x' = \frac{x - \text{min}(x)}{\text{max}(x)-\text{min}(x)}$$

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
x1 = X_tr.values #returns a numpy array
x2 = X_te.values
y1 = y_tr.values
y2 = y_te.values
min_max_scaler = MinMaxScaler()
X_train = min_max_scaler.fit_transform(x1.reshape(-1,1))
X_test = min_max_scaler.fit_transform(x2.reshape(-1,1))
y_train = min_max_scaler.fit_transform(y1.reshape(-1,1))
y_test = min_max_scaler.fit_transform(y2.reshape(-1,1))
```

Ορισμός συνάρτησης MSE:
$$ MSE = \frac{1}{n}\sum_{i=0}^n (y_i - \hat{y}_i )^2$$ 

```python
def MSE(y_pred,y_actual):
    if(len(y_pred) != len(y_actual)):
        print('number of predicted and actual values is not the same!')
        return
    n = len(y_pred)
    return (1/n)*sum((y_pred-y_actual)**2)      
```

Εδώ εφαρμόζουμε απλό Linear Regression στο dataset 1a με τη βοήθεια του **Scikit learn**, ώστε να έχουμε μια εικόνα ποιες είναι οι 'σωστές' τιμές για τις παραμέτρους $\alpha,\beta$ καθώς και να βεβαιωθούμε ότι η συνάρτηση μας για το MSE βρίσκει σωστά αποτελέσματα.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
reg = LinearRegression().fit(X_train, y_train)
print('coefficients are a= {} and b={}.'.format(reg.coef_,reg.intercept_))
y_pred_SL = reg.predict(X_test)
print('score = ',reg.score(y_pred_SL,y_test))
print('MSE = ',MSE(y_pred_SL,y_test))
print('MSE from scikit for verification = ',mean_squared_error(y_pred_SL,y_test))
```

Επομένως, παρατηρούμε ότι η συνάρτησή μας υπολογίζει σωστά το MSE, καθώς και ότι οι τιμές για του συντελεστές αναμένουμε να είναι της τάξης του: $\alpha = 0.94$ και $\beta = 0.006$.

#### Stochastic Gradient Descent Algorithm

Θα προσπαθήσω να γράψω τα μαθηματικά του Gradient Descent με σκοπό να τα καταλάβω ο ίδιος καλύτερα.\
Αν η ευθεία ελαχίστων τετραγώνων που προσεγγίζει τα σημεία μου κάποια στιγμή είναι η: 
$$ \hat{y_i} = \alpha x_i + \beta  \ (+ \varepsilon)$$ 
τότε για το MSE ισχύει:
$$ MSE = \frac{1}{n}\sum_{i=0}^n (y_i - (\alpha\hat{x_i} + \beta) )^2$$ 
Οπότε, για να το ελαχιστοποιήσω, πρέπει να υπολογίσω τις μερικές παραγώγους προς τα $\alpha,\beta$. Έχουμε:
$$D_{\alpha} = \frac{\partial(MSE)}{\partial \alpha} = \frac{-2x_i}{n}\sum_{i=0}^n (y_i - (\alpha\hat{x_i} + \beta) )$$
$$D_{\beta} = \frac{\partial(MSE)}{\partial \beta} = \frac{-2}{n}\sum_{i=0}^n (y_i - (\alpha\hat{x_i} + \beta) )$$
Άρα, αν το learning rate είναι το $\eta$, οι κανόνες ανανέωσης είναι:
$$ \alpha' = \alpha - \eta\cdot D_{\alpha}$$
και:
$$ \beta' = \beta - \eta\cdot D_{\beta}$$

Σε αυτό που διαφοροποιείται ο **Stochastic Gradient Descent** είναι ότι για κάθε επανάληψη παίρνει ένα τυχαίο δείγμα από το training set.  
Επομένως, εφόσον είναι online μέθοδος, δεν έχουμε άθροισμα, και ο κανόνας ανανέωσης για κάθε επανάληψη (δηλαδή για ένα δείγμα τη φορά) είναι:
$$ \alpha' = \alpha - 2\eta\cdot x_i (\hat{y_i}-(\alpha\cdot x_i+\beta))$$
και:
$$ \beta' = \beta - 2\eta\cdot (\hat{y_i}-(\alpha\cdot x_i+\beta))$$


```python
def SGD(X, y, xtest,ytest,lr=0.05, epochs=100):
        
    '''
    Online Stochastic Gradient Descent (mini-batch-size = 1)
    with plots
    '''
    a,b = 0.,0.
    log, mse,mse_te = [], [], []
    fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,8))
    N = len(X)
    for k in range(epochs):
        Xs,ys = shuffle(X,y,random_state=24)
        Xs = Xs.reshape(-1,)
        ys = ys.reshape(-1,)
        for i,j in zip(Xs,ys):

            f = j - (a*i+b)
            #update weights
            a -= lr*(-2*i*f)
            b -= lr*(-2*f)
        mse.append(MSE(y,a*X+b))
        mse_te.append(MSE(ytest,a*xtest+b))
        
        log.append((a,b))
    
    #PLOTS AND OUTPUT
    ep = [p for p in range(0,epochs)]
    ax2.plot(ep,mse,'b--',label='Train Set MSE')
    ax2.plot(ep,mse_te,'r--',label='Test Set MSE')
    ax2.legend()
    ax2.margins(0.05)
    y_pred = a*xtest+b 
    message = 'coefficients are a = {} and b = {}, with MSE = {}'.format(a,b,MSE(y_pred,ytest))
    axes = plt.gca()
    ax2.set_title('MSE vs Number of Epochs')
    ax1.plot(X, a*X+b, 'r--')
    ax1.scatter(X,y)
    ax1.set_title('Training Data Scatter and Line Fit (a*X_train+b)')
    ax1.margins(0.05)
    ax3.plot(xtest, a*xtest+b, 'r--')
    ax3.scatter(xtest,ytest)
    ax3.set_title('Test Data Scatter and Line Fit (a*X_test+b)')
    ax3.margins(0.05)

    plt.show()
    return message
```

```python
%%time
SGD(X_train,y_train,X_test,y_test,lr=0.015,epochs=17)
```

Κάνοντας διάφορα πειράματα ως προς την εκλογή του κατάλληλου learning rate και του αριθμού των epochs, καταλήξαμε στο ότι με Learning Rate = 0.015 σε 15-20 epochs επιτυγχάνουμε καλή σύγκλιση. Δεδομένου ότι με σκοπό τη σύγκλιση αυτά τα δύο μεγέθη είναι αντιστρόφως ανάλογα, με πιο ενδελεχή έλεγχο ίσως να βρίσκαμε και καταλληλότερες τιμές, όμως δεν υπάρχει λόγος καθώς έχουμε καλή σύγκλιση σε ελάχιστο χρόνο $<1 sec$.  
**Σχόλιο**: μπορεί να φαίνεται ότι το MSE στο train set είναι πολύ χαμηλότερο απ'ότι του test set όπως η κλίμακα είναι πολύ μικρή επομένως οι διαφορές φαίνονται μεγαλύτερες.
