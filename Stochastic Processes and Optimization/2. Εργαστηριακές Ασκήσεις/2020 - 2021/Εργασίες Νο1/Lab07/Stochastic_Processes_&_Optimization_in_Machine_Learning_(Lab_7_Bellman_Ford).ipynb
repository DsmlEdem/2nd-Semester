 - >  Να περιγράψετε, συνοπτικά, τη λειτουργία του αλγορίθμου Bellman-Ford.
 
Ο αλγόριθμος Bellman-Ford δέχεται ως input έναν γράφο και έναν κόμβο ως αφετηρία, και δίνει output την συντομότερη διαδρομή όλων των κόμβων από την αφετηρία.  
Ο αλγόριθμος είναι αλγόριθμος δυναμικού προγραμματισμού, αρχικοποιεί την απόσταση από την κορυφή της αφετηρίας ως 0 και τις αποστάσεις προς όλους τους άλλους κόμβους στο άπειρο. Έπειτα σαρώνει όλες τις ακμές και αν η απόσταση είναι εφικτό να γίνει μικρότερη, την ενημερώνει στην χαμηλότερη τιμή. Για κάθε επανάληψη n που σαρώνει τις ακμές, ο αλγόριθμος Bellman-Ford βρίσκει όλες τις ελάχιστες διαδρομές με μέγιστο μήκος n ακμών. Αν το πλήθος των κόμβων είναι V και η μεγαλύτερη δυνατή διαδρομή χωρίς κύκλο μπορεί να είναι V-1 ακμές, τότε οι ακμές πρέπει να σαρωθούν V-1 φορές για να εξασφαλιστεί ότι για κάθε κόμβο έχουμε την συντομότερη διαδρομή. Στο τέλος γίνεται μια τελική σάρωση όλων των άκρων, και αν υπάρξει ενημέρωση για κάποια απόσταση έχει βρεθεί μια διαδρομή μήκους V άκρων που μπορεί να συμβεί εάν υπάρχει τουλάχιστον ένας αρνητικός κύκλος στο διάγραμμα. 

 - > Να εφαρμόσετε τον αλγόριθμο Bellman-Ford στο γράφο του παραπάνω σχήματος για τις ακμές που συμβολίζονται με μαύρο χρώμα και συνεχή γραμμή. Ως πηγή του αλγορίθμου να θεωρήσετε το AS#2. Ποιο είναι το συνολικό κόστος της διαδρομής από το AS#2 στο AS#11; Ποιο είναι το μονοπάτι του δέντρου Bellman-Ford από το AS#2 στο AS#11;

```python
import bellmanford as bf
import networkx as nx  
import matplotlib.pyplot as plt
```

Αρχικοποιούμε ένα γράφο με 13 κόμβους, χωρίς τα edges, δηλαδή τις ενώσεις με τα βάρη τους.

```python
G=nx.Graph()#create empty Graph
nodes=["AS#1","AS#2","AS#3","AS#4","AS#5","AS#6","AS#7","AS#8","AS#9","AS#10","AS#11","AS#13"]#nodeslist
G.add_nodes_from(nodes)#add nodes
```

Επίσης μια συνάρτηση που να οπτικοποιεί τον γράφο που δημιουργούμε.

```python
def printGraph(G):
    fig = plt.subplots(figsize=(8,8))
    # Create positions of all nodes and save them
    pos = nx.spring_layout(G)
    # Draw the graph according to node positions
    nx.draw(G, pos, with_labels=True)
    # Create edge labels
    labels = {e: G.get_edge_data(*e)['weight'] for e in G.edges}
    # Draw edge labels according to node positions
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
    plt.show()
printGraph(G)
```

Ας προσθέσουμε λοιπόν και τις ακμές.

```python
G.add_edge("AS#1", "AS#3",weight=4)
G.add_edge("AS#2", "AS#3",weight=4)
G.add_edge("AS#3", "AS#5",weight=2)
G.add_edge("AS#4", "AS#5",weight=20)
G.add_edge("AS#5", "AS#6",weight=5)
G.add_edge("AS#5", "AS#7",weight=2)
G.add_edge("AS#6", "AS#7",weight=4)
G.add_edge("AS#7", "AS#8",weight=3)
G.add_edge("AS#8", "AS#9",weight=7)
G.add_edge("AS#8", "AS#13",weight=4)
G.add_edge("AS#13", "AS#9",weight=5)
G.add_edge("AS#9", "AS#10",weight=3)
G.add_edge("AS#9", "AS#11",weight=4)
```

```python
printGraph(G)
```

```python
path_length, path_nodes, negative_cycle = bf.bellman_ford(G, source='AS#2', target='AS#11')
print("Is there a negative cycle? {0}".format(negative_cycle))
print("Shortest path length: {0}".format(path_length))
print("Shortest path: {0}".format(path_nodes))
```

Επομένως το συντομότερο μοντοπάτι από το AS#2 στο AS#11 είναι το: 
$$\boldsymbol{AS\#2 \rightarrow AS\#3 \rightarrow AS\#5 \rightarrow AS\#7 \rightarrow AS\#8 \rightarrow AS\#9 \rightarrow AS\#11}$$  
με **συνολικό κόστος 22**.

- > Το AS#2 εξετάζει την προσθήκη μιας ζεύξης είτε με το AS#4 ή με το AS#1. Οι ζεύξεις αυτές φαίνονται στον παραπάνω γράφο με μαύρες διακεκομμένες γραμμές. Γιατί πιστεύετε ότι χρειάζεται το AS#2 την επιπλέον αυτή ζεύξη; Ποια από τις παραπάνω ζεύξεις θα πρέπει να προτιμήσει το AS#2 για να είναι μικρότερο το κόστος της διαδρομής από το AS#2 μέχρι το AS#11;

Ο κόμβος $AS\#2$ χρειάζεται μια επιπλέον ζεύξη είτε με το $AS\#1$ είτε με το $AS\#4$, διότι αν η ζεύξη $AS\#2\leftrightarrow AS\#3$ πάθει βλάβη, τότε ο κόμβος $AS\#2$ αποκόπτεται από τον υπόλοιπο γράφο αφού δεν επικοινωνεί με κανέναν άλλο.   
Για να είναι μικρότερο το κόστος της διαδρομής από το $AS\#2$ στο $AS\#11$ τότε θα πρέπει να προτιμηθεί η ζεύξη με το $AS\#1$, αφού έχει κόστος 16 μέχρι τον $AS\#5$ ενώ η ζεύξη με το $AS\#4$ έχει κόστος 24.    
Παρ'όλα αυτά, αν δεν μας ενδιέφερε το κόστος και θέλαμε να είναι πιο fault proof ο γράφος, η επιλογή να προστεθεί η ζεύξη με το $AS\#4$ θα ήταν καλύτερη, γιατί τόσο ο κόμβος $AS\#2$ όσο και ο $AS\#1$ συνδέονται με τον $AS\#3$, επομένως αν πάθαινε κάτι ο κόμβος αυτός οι άλλοι δύο θα έμεναν αποκομμένοι από τον γράφο, ενώ η ζεύξη $AS\#2\leftrightarrow AS\#4$ δίνει εναλλακτικό μονοπάτι και επικοινωνεί με τον υπόλοιπο γράφο. Έτσι δηλαδή επιτυγχάνουμε δυο διαφορετικές διαδρομές προς τον $AS\#5$.

- > Προσθέστε την κατάλληλη ζεύξη, όπως την προσδιορίσατε στο προηγούμενο ερώτημα. Έστω ότι η ζεύξη AS #2 – AS #3 παθαίνει μια βλάβη. Ποιο είναι, τώρα, το κόστος της διαδρομής από το AS #2 μέχρι το AS #11; Πόσο επιβαρύνθηκε σε σχέση με το δεύτερο ερώτημα;

```python
G.remove_edge('AS#2','AS#3')
G.add_edge('AS#2','AS#1',weight=10)
```

```python
printGraph(G)
```

```python
path_length, path_nodes, negative_cycle = bf.bellman_ford(G, source='AS#2', target='AS#11')
print("Is there a negative cycle? {0}".format(negative_cycle))
print("Shortest path length: {0}".format(path_length))
print("Shortest path: {0}".format(path_nodes))
```

**Παρατηρούμε τώρα ότι το κόστος της διαδρομής είναι 32 αντί για 22 που ήταν προηγουμένως, επομένως επιβαρύνθηκε κατά 10 μονάδες, ή $45.5\%$.**

- > Έστω ότι αποκαθίσταται η βλάβη. Το AS #12 προσφέρεται να ενώσει τα AS #10 και AS #11 απ’ ευθείας στο AS #3. Το AS #12 μαζί με τις νέες ακμές που θα προστεθούν φαίνονται με κόκκινο χρώμα στον παραπάνω γράφο. Πόσο θα βελτιωθεί το συνολικό κόστος AS #2 – AS #11; Το συγκεκριμένο ερώτημα προσπαθεί να προσομοιώσει τη λειτουργία του GR-IX.
```python
G.add_edge('AS#2','AS#3',weight=4)
G.add_node('AS#12')
G.add_edge('AS#3','AS#12',weight=2)
G.add_edge('AS#12','AS#10',weight=2)
G.add_edge('AS#12','AS#11',weight=4)
```
