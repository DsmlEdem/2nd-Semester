```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.datasets import make_blobs
from sklearn.svm import SVC # "Support vector classifier"
from sklearn.datasets import make_circles
from ipywidgets import interact, fixed
from mpl_toolkits import mplot3d
import seaborn as sns; sns.set()
```

Το παρακάτω τμήμα κώδικα δημιουργεί και εμφανίζει δείγματα από 2 clusters με μία δεδομένη τυπική απόκλιση.

```python
X, y = make_blobs(n_samples=50, centers=2,
                  random_state=0, cluster_std=0.60)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn');
```

Γίνεται εύκολα αντιληπτό ότι οι δύο κατηγορίες μπορούν να διαχωριστούν με ευθείες γραμμές. Στο παρακάτω σχήμα φαίνονται 3 τέτοιες ενδεικτικές γραμμές.

<li><b>Ερώτηση 2:</b> Γιατί αυτή η μέθοδος διαχωρισμού των κλάσεων ενός ταξινομητή μπορεί να οδηγήσει σε μη επιθυμητά αποτελέσματα;</li>
</ul>

<font color='#486393'><i>Ο κώδικας του επόμενου κελιού υπέστη μια ελαφρά επεξεργασία, προκειμένου να είναι εφικτή η αναφοράς στις τρεις διαφορετικές ευθείες (εισήχθη χρωματικός κώδικας).</i></font>

- <font color='#486393'>Όπως γίνεται εύκολα αντιληπτό, το νέο δειγματικό σημείο θα ταξινομηθεί στην ίδια κατηγορία με όλα τα κόκκινα σημεία στην περίπτωση που επιλεχθεί η μπλε ή η πράσινη ευθεία ως διαχωριστική. Στην περίπτωση που η μαύρη ευθεία επιλεχθεί ως διαχωριστική, το νέο δειγματικό σημείο θα ταξινομηθεί στην ίδια κατηγορία με όλα τα κίτρινα σημεία.</font>

- <font color='#486393'>Παρότι όλες οι προκύπτουσες ευθείες επιτυγχάνουν να διαχωρίσουν σωστά τα δεδομένα στις δύο κατηγορίες, δεν είναι όλες τους ισοδύναμες σε ό,τι έχει να κάνει με τη δυνατότητα γενίκευσής τους. Για παράδειγμα, μια ευθεία η οποία εμφανίζει πολύ μικρότερη μέση απόσταση ανά σημείο από τη μία κατηγορία σε σχέση με την άλλη έχει μια εγγενή προτίμηση στην ταξινόμηση νέων σημείων στην κατηγορία από την οποία απέχει τη μεγαλύτερη μέση απόσταση ανά σημείο. Αυτή είναι, εξάλλου, η αφορμή για την ανάπτυξη των Support Vector Machines (SVMs), τα οποία αποτελούν το κεντρικό αντικείμενο μελέτης του παρόντος notebook: η «εξισορρόπηση» μεταξύ των αποστάσεων των σημείων κάθε κατηγορίας από την ευθεία διαχωρισμού.</font>

Στην περίπτωση του ταξινομητή SVM το πρόβλημα αυτό επιλύεται με τον παρακάτω τρόπο: Αντί να τραβήξουμε μία γραμμή μηδενικού πλάτους ανάμεσα στις δύο κλάσεις, τραβάμε μία γραμμή με περιθώρια (margins) στις δύο πλευρές της. Η μέθοδος αυτή για τις γραμμές του προηγούμενου προβλήματος φαίνεται στην παρακάτω εικόνα. 
<ul>

- <font color='#486393'>Σε ό,τι έχει να κάνει με τις περιθωριακές ευθείες (σε υψηλότερες διαστάσεις γίνεται λόγος για περιθωριακά υπερεπίπεδα - marginal hyperplanes), αυτές θα πρέπει καταρχάς να ισαπέχουν από την ευθεία διαχωρισμού (και προφανώς να είναι παράλληλες προς αυτήν). Επιπροσθέτως, οι περιθωριακές ευθείες θα πρέπει να απέχουν όσο το δυνατό μεγαλύτερη απόσταση από την ευθεία διαχωρισμού, χωρίς όμως να αφήνουν σημεία να παρεμβάλονται ανάμεσά σε αυτές και στην ευθεία διαχωρισμού (το πολύ κάποια σημεία να εφάπτονται σε αυτές). Οι συνθήκες αυτές ορίζουν πλήρως τις περιθωριακές ευθείες σε κάθε τέτοιο πρόβλημα με δεδομένα που ανήκουν σε γραμμικά διαχωρίσιμες κατηγορίες.</font>
    
- <font color='#486393'>Βάσει των παραπάνω, η βέλτιστη διαχωριστική ευθεία (εν γένει υπερεπιφάνεια) είναι αυτή για την οποία η απόστασή της από τις περιθωριακές ευθείες (εν γένει περιθωριακά υπερεπίπεδα) είναι η μέγιστη δυνατή.</font>

```python
xfit = np.linspace(-1, 3.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')

for m, b, col, d in [(1, 0.65, 'green', 0.33), (0.5, 1.6, 'blue', 0.55), (-0.2, 2.9, 'black', 0.2)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, '-', color=col)
    plt.fill_between(xfit, yfit - d, yfit + d, edgecolor='none',
                     color='#AAAAAA', alpha=0.4)

plt.xlim(-1, 3.5);
```

### Εκπαίδευση του Μοντέλου

Χρησιμοποιώντας το πακέτο Scikit-Learn θα εκπαιδεύσουμε τον αλγόριθμο SVM πάνω στα δεδομένα του προηγούμενου προβλήματος, χρησιμοποιώντας γραμμικό πυρήνα.

```python
model = SVC(kernel='linear', C=1E10)
model.fit(X, y)
```

Με την παρακάτω μέθοδο θα εμφανιστεί στο σχήμα που ακολουθεί η βέλτιστη γραμμή διαχωρισμού των δύο κατηγοριών, καθώς και τα περιθώριά της. Τα training examples που ακουμπούν πάνω στα όρια ονομάζονται Support Vectors.

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plot_svc_decision_function(model);
```

Στο ακόλουθο σχήμα, η διαδικασία επαναλαμβάνεται για δύο datasets που περιλαμβάνουν 60 και 120 training examples αντίστοιχα. 
<ul>
<li><b>Ερώτηση 5:</b> Τι παρατηρείτε σχετικά με τη βέλτιστη γραμμή που διαχωρίζει τις δύο κατηγορίες και τα περιθώριά της;</li>
<li><b>Ερώτηση 6:</b>Τι αρκεί να κάνει ο αλγόριθμος για να ταξινομήσει νέα δεδομένα που θα λάβει στην είσοδό του;</li>
</ul>

- <font color='#486393'>Η βέλτιστη ευθεία διαχωρισμού ικανοποιεί όλες τις προαναφερθείσες συνθήκες. Όπως φαίνεται και από το σχήμα, οι αντίστοιχες περιθωριακές ευθείες εφάπτονται σε ορισμένα σημεία, κάποια εκ των οποίων (στην προκείμενη περίπτωση όλα, το οποίο όμως δεν ισχύει στη γενικότητα, αφού είναι δυνατό σημεία που ανήκουν στις περιθωριακές ευθείες να μην είναι διανύσματα υποστήριξης - βλ. <i>Foundations of Machine Learning, M. Mohri, A. Rostamizadeh, A. Talwalkar)</i> αποτελούν τα λεγόμενα διανύσματα υποστήριξης (support vectors). Η ονομασία τους οφείλεται στο γεγονός πως αυτά είναι τα μοναδικά σημεία τα οποία καθορίζουν τη βέλτιστη διαχωριστική ευθεία (και κατ' επέκταση τις περιθωριακές ευθείες) η οποία είναι ανεξάρτητη από όλα τα υπόλοιπα σημεία του δείγματος.</font>
    
- <font color='#486393'>Εάν η διαχωριστική ευθεία περιγράφεται από την εξίσωση $g\left(\mathbf{x}\right) = 0$, τότε κάθε νέο δειγματικό σημείο, $\mathbf{x}^\prime$, ταξινομείται στην ίδια κατηγορία με τα κόκκινα σημεία εφόσον $g\left(\mathbf{x}^\prime\right) > 0$. Από την άλλη, εάν $g\left(\mathbf{x}^\prime\right) < 0$, τότε το νέο δειγματικό σημείο ταξινομείται στην ίδια κατηγορία με τα κίτρινα σημεία. Στη σπάνια περίπτωση κατά την οποία προκύπτει $g\left(\mathbf{x}^\prime\right) = 0$, η ταξινόμηση είτε δεν πραγματοποιείται σε καμία κατηγορία, είτε πραγματοποιείται τυχαία.</font>

```python
def plot_svm(N=10, ax=None):
    X, y = make_blobs(n_samples=200, centers=2,
                      random_state=0, cluster_std=0.60)
    X = X[:N]
    y = y[:N]
    model = SVC(kernel='linear', C=1E10)
    model.fit(X, y)
    
    ax = ax or plt.gca()
    ax.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
    ax.set_xlim(-1, 4)
    ax.set_ylim(-1, 6)
    plot_svc_decision_function(model, ax)

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)
for axi, N in zip(ax, [60, 120]):
    plot_svm(N, axi)
    axi.set_title('N = {0}'.format(N))
```

### Μη Γραμμικά Διαχωρίσιμες Κατηγορίες: Kernel SVM

Στην περίπτωση που οι κατηγορίες του προβλήματος δεν είναι γραμμικά διαχωρίσιμες, ο αλγόριθμος SVM χρησιμοποιεί μη γραμμικούς πυρήνες (kernels) για να προβάλει τα δεδομένα εκπαίδευσης σε μεγαλύτερη διάσταση όπου μπορεί να τα καταστήσει γραμμικά διαχωρίσιμα. Μία τέτοια περίπτωση δεδομένων εκπαίδευσης παρουσιάζεται στο επόμενο σχήμα.

```python
X, y = make_circles(100, factor=.1, noise=.1)

clf = SVC(kernel='linear').fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plot_svc_decision_function(clf, plot_support=False);
```

Είναι φανερό ότι τα δεδομένα εκπαίδευσης δεν μπορούν να διαχωριστούν γραμμικά. Στο επόμενο σχήμα φαίνεται η προβολή τους στις τρεις διαστάσεις με τη χρήση ενός μη γραμμικού πυρήνα. 
<ul>
<li><b>Ερώτηση 7:</b> Τι είδους πυρήνας χρησιμοποιήθηκε;</li>
<li><b>Ερώτηση 8:</b> Δώστε ένα παράδειγμα γραμμικού ορίου που θα μπορούσε να διαχωρίσει τις δύο κατηγορίες.</li>
</ul>

- <font color='#486393'>Ο μη γραμμικός πυρήνας που χρησιμοποιήθηκε ανήκει στην ευρύτερη κατηγορία των ακτινικών συναρτήσεων βάσης (radial basis functions) οι οποίες αναλύθηκαν εκτενώς στο άλλο notebook της παρούσας εργαστηριακής άσκησης. Πιο συγκεκριμένα, είναι ένας Γκαουσιανός πυρήνας με $\sigma^2 = 0.5$ και κέντρο την αρχή των αξόνων:</font>

<font color='#486393'>$$ \phi\left(\mathbf{x}\right) = \exp\left(-||\mathbf{x}||^2\right) $$</font>
    
- <font color='#486393'>Στα πλαίσια της απεικόνισης των δεδομένων στις τρεις διαστάσεις, ο διαχωρισμός τους πλέον καθίσταται εφικτός μέσω ενός επιπέδου που να είναι παράλληλο στο επίπεδο που ορίζουν οι άξονες $x$ και $y$. Αν και υπάρχουν άπειρα τέτοια επίπεδα, ένα παράδειγμα είναι το επίπεδο $r = 0.61$ (με $r$ να είναι ο άξονας που είναι ορθοκανονικός ως προς τους $x$ και $y$).</font>

```python
r = np.exp(-(X ** 2).sum(1))
```

```python
def plot_3D(elev=30, azim=30, X=X, y=y):
    ax = plt.subplot(projection='3d')
    ax.scatter3D(X[:, 0], X[:, 1], r, c=y, s=50, cmap='autumn')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('r')

interact(plot_3D, elev=[-90, 90], azip=(-180, 180),
         X=fixed(X), y=fixed(y));
```

Η εύρεση του πυρήνα που μπορεί να διαχωρίσει με τον καλύτερο τρόπο τα δεδομένα απαιτεί, συνήθως, πολλούς υπολογισμούς. 
<ul>
<li><b>Ερώτηση 9:</b> Ποια μέθοδος μπορεί να βοηθήσει σε αυτό και πώς; (hint: <a href="https://en.wikipedia.org/wiki/Kernel_trick">Kernel Trick</a>)</li>
</ul>

- <font color='#486393'>Η κεντρική ιδέα είναι πως, παρότι η απεικόνιση των δεδομένων, $\mathbf{x}$, γίνεται σε έναν νέο χώρο (όπου αυτά είναι γραμμικά διαχωρίσιμα) μέσω κάποιας συνάρτησης $\phi\left(\mathbf{x}\right)$, ο προσδιορισμός αυτής της συνάρτησης δεν είναι απαραίτητος προκειμένου να βρεθεί η εξίσωση της διαχωριστικής υπερεπιφάνειας. Ο λόγος για αυτό είναι πως η εξίσωση της υπερεπιφάνειας είναι $g\left(\mathbf{x}\right) = 0$, με</font>

<font color='#486393'>$$ g\left(\mathbf{x}\right) = w_0 + \sum_{i=1}^{N}\lambda_i d_i \phi\left(\mathbf{x}_i\right)^\top\phi\left(\mathbf{x}\right),$$</font>

<font color='#486393'>όπου ο δείκτης $i$ τρέχει στα δεδομένα εκπαίδευσης, οι ποσότητες $\lambda_i$ είναι οι αντίστοιχοι πολλαπλασιαστές Lagrange (μη μηδενικοί μόνο για τα διανύσματα υποστήριξης), οι ποσότητες $d_i$ είναι ίσες με $\pm 1$ ανάλογα με την κατηγορία στην οποία ανήκει το αντίστοιχο σημείο και $w_0$ είναι μια σταθερά. Όπως γίνεται εμφανές στην άνω σχέση, η συνάρτηση $\phi\left(\mathbf{x}\right)$ δεν εμφανίζεται μόνη της, αλλά ως ένα εσωτερικό γινόμενο $\phi\left(\mathbf{x}_i\right)^\top\phi\left(\mathbf{x}\right)$. Το συγκεκριμένο εσωτερικό γινόμενο μπορεί να αντιστοιχιστεί σε μια συνάρτηση πυρήνα $K\left(\mathbf{x}_i,\mathbf{x}\right)$ (υπό προϋποθέσεις που τίθενται από το Θεώρημα του Mercer), επομένως το πρόβλημα μπορεί να αναχθεί στον προσδιορισμό αυτής της συνάρτησης πυρήνα. Η αναγωγή αυτή είναι πολύ υψηλής σημασίας, αφού η αντίστοιχη υπολογιστική πολυπλοκότητα είναι σημαντικά μικρότερη - συγκεκριμένα είναι $\mathcal{O}\left(N\right)$, με $N$ το συνολικό πλήθος σημείων εκπαίδευσης, έναντι της $\mathcal{O}\left(N^2\right)$ που αντιστοιχεί στην εύρεση της $\phi$.</font>

Παρακάτω γίνεται η εκπαίδευση του αλγορίθμου SVM με μη γραμμικό πυρήνα.

```python
clf = SVC(kernel='rbf', C=1E6)
clf.fit(X, y)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plot_svc_decision_function(clf)
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
            s=300, lw=1, facecolors='none');
```
