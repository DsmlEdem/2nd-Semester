### Απαντήσεις

- <font color='#486393'>Ο αλγόριθμος Bellman-Ford αφορά στην εύρεση της συντομότερης διαδρομής από έναν πηγαίο κόμβο ενός γράφου προς όλους τους υπόλοιπους κόμβους του γράφου αυτού.  Συγκεκριμένα, ο αλγόριθμος αρχικοποιείται θέτοντας την απόσταση κάθε άλλου κόμβου από τον πηγαίο κόμβο ίση με κάποια υπερεκτιμημένη προσέγγιση της ελάχιστης απόστασης (σε περιπτώσεις όπου δε γίνεται ή δεν είναι επιθυμητή μια πεπερασμένη υπερεκτίμηση, όλες οι αποστάσεις από τον πηγαίο κόμβο τίθενται ίσες με το άπειρο). Σε κάθε του επανάληψη, έστω υπ' αριθμόν $t$, ο αλγόριθμος σαρώνει όλες τις ακμές του γράφου και εκτιμά εκ νέου όλες τις ελάχιστες αποστάσεις με μέγιστο μήκος $t$ ακμών. Το συνολικό πλήθος των επαναλήψεων είναι ίσο με $|V|-1$, όπου $|V|$ το πλήθος των κόμβων του γράφου, αφού αυτό είναι το μέγιστο δυνατό πλήθος βημάτων για τη μετάβαση από έναν κόμβο σε έναν άλλο. Έτσι, κάθε απόσταση ανανεώνεται εκ νέου σε κάθε επανάληψη, αποκτώντας με το πέρας της επαναληπτικής διαδικασίας την πραγματική ελάχιστη τιμή της.</font>

- <font color='#486393'>Για την προσέγγιση του προβλήματος θα αξιοποιηθεί η βιβλιοθήκη [bellmanford](https://pypi.org/project/bellmanford/) για την εφαρμογή του αλγορίθμου Bellman-Ford, καθώς και η [networkx](https://networkx.org/) για την απεικόνιση των σχετικών γράφων.</font>

```python
import bellmanford as bf
import networkx as nx

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = "darkgrid") # καθαρά αισθητικοί λόγοι

# Συνάρτηση για την απεικόνιση των γράφων
def DrawGraph(Graph):
    fig = plt.subplots(1,1,figsize=(10,10))
    positions = nx.spring_layout(Graph)
    edge_color, node_color = '#818fa3', '#818fa3'
    
    nx.draw_networkx(Graph, positions, with_labels=True, edge_color=edge_color,
                     font_color='#ffffff',node_color=node_color, node_size=1600)
    #nx.draw_networkx(Graph, with_labels=True, node_color=node_color, edge_color=edge_color)
    
    labels = {edge: Graph.get_edge_data(*edge)['weight'] for edge in Graph.edges}
    nx.draw_networkx_edge_labels(Graph, positions, edge_labels=labels)
    plt.show()
    
# Συνάρτηση για την κατασκευή ενός γράφου
def CreateGraph(nodes,edges):
    Graph = nx.Graph()
    
    Graph.add_nodes_from(nodes)
    Graph.add_weighted_edges_from(edges)
    return Graph
```

```python
Nodes_list = ["AS#1","AS#2","AS#3","AS#4","AS#5","AS#6","AS#7","AS#8","AS#9","AS#10","AS#11","AS#13"]
Edges_list = [("AS#1","AS#3",4), ("AS#2","AS#3",4), ("AS#3","AS#5",2), ("AS#5","AS#4",20), ("AS#5","AS#6",5),
              ("AS#5","AS#7",2), ("AS#6","AS#7",4), ("AS#7","AS#8",3), ("AS#8","AS#13",4), ("AS#8","AS#9",7),
              ("AS#13","AS#9",5), ("AS#9","AS#10",3), ("AS#9","AS#11",4)]

Graph = CreateGraph(Nodes_list,Edges_list)
cost, path, neg_c = bf.bellman_ford(Graph, source='AS#2', target='AS#11')

print("Η ακόλουθη εικόνα απεικονίζει τον υπό μελέτη γράφο.")
DrawGraph(Graph)

print(f"Βάσει του αλγορίθμου Bellman-Ford, το συνολικό κόστος της διαδρομής είναι {cost}, ενώ το βέλτιστο μονοπάτι είναι το:")
print(f"{path}")
```

<center><font color='#486393'><b>AS#2 $\rightarrow$ AS#3 $\rightarrow$ AS#5 $\rightarrow$ AS#7 $\rightarrow$ AS#8 $\rightarrow$ AS#9 $\rightarrow$ AS#11</b></font></center>

- <font color='#486393'>Σε ό,τι έχει να κάνει με την προσθήκη μιας επιπλέον ζεύξης στον κόμβο AS#2, αυτή κρίνεται απαραίτητη δεδομένου πως ο κόμβος AS#2 είναι ο πηγαίος κόμβος του γράφου στα πλαίσια του αλγορίθμου Bellman-Ford. Η μόνη ζεύξη που ενώνει τον κόμβο αυτό με τον υπόλοιπο γράφο είναι η AS#2 $\leftrightarrow$ AS#3 και σε περίπτωση βλάβης της ο κόμβος AS#2 απομονώνεται από τον υπόλοιπο γράφο. Σε ό,τι αφορά το κατά πόσο η ζεύξη AS#2 $\leftrightarrow$ AS#1 ή η ζεύξη AS#2 $\leftrightarrow$ AS#4 είναι προτιμότερη προκειμένου να ελαχιστοποιείται το συνολικό κόστος διαδρομής, το παρακάτω κομμάτι κώδικα φροντίζει να υπολογίσει τα δύο κόστη και να αποφανθεί ως προς το ερώτημα αυτό.</font>

```python
two_to_one_edges, two_to_four_edges = Edges_list.copy(), Edges_list.copy()
two_to_one_edges.append(("AS#2","AS#1",10))
two_to_four_edges.append(("AS#2","AS#4",4))

Graph_tto = CreateGraph(Nodes_list,two_to_one_edges)
cost_tto, path_tto, neg_c = bf.bellman_ford(Graph_tto, source='AS#2', target='AS#11')
print(f'Μετά την προσθήκη της ζεύξης AS#2 <-> AS#1, το συνολικό κόστος της βέλτιστης διαδρομής είναι {cost_tto}.')

Graph_ttf = CreateGraph(Nodes_list,two_to_four_edges)
cost_ttf, path_ttf, neg_c = bf.bellman_ford(Graph_ttf, source='AS#2', target='AS#11')
print(f'Μετά την προσθήκη της ζεύξης AS#2 <-> AS#4, το συνολικό κόστος της βέλτιστης διαδρομής είναι {cost_ttf}.')
```

<font color='#486393'>Παρατηρούμε πως η προσθήκη κάθε ζεύξης αυτής καθ' αυτής, χωρίς περαιτέρω μεταβολές του γράφου, δε μεταβάλει το συνολικό κόστος της βέλτιστης διαδρομής. Αυτό οφείλεται στο γεγονός πως η βέλτιστη διαδρομή παραμένει ίδια με αυτήν που υπολογίστηκε στο προηγούμενο ερώτημα, παρά την προσθήκη της ζεύξης. Προκειμένου, λοιπόν, να ελέγξουμε ποια ζεύξη είναι πιο κοστοβόρα θα πρέπει αυτό να γίνει στο σενάριο όπου πραγματοποιείται η συνθήκη που καθιστά την προσθήκη αυτή αναγκαία, δηλαδή στο σενάριο όπου η ζεύξη AS#2 $\leftrightarrow$ AS#3 παθαίνει βλάβη.</font>

```python
two_to_one_edges.remove(("AS#2","AS#3",4))
two_to_four_edges.remove(("AS#2","AS#3",4))

Graph_tto = CreateGraph(Nodes_list,two_to_one_edges)
cost_tto, path_tto, neg_c = bf.bellman_ford(Graph_tto, source='AS#2', target='AS#11')
print(f'Στην περίπτωση βλάβης της ζεύξης AS#2 <-> AS#3, η προσθήκη της ζεύξης AS#2 <-> AS#1 οδηγεί σε συνολικό κόστος της νέας βέλτιστης διαδρομής ίσο με {cost_tto}.\n')

Graph_ttf = CreateGraph(Nodes_list,two_to_four_edges)
cost_ttf, path_ttf, neg_c = bf.bellman_ford(Graph_ttf, source='AS#2', target='AS#11')
print(f'Στην περίπτωση βλάβης της ζεύξης AS#2 <-> AS#3, η προσθήκη της ζεύξης AS#2 <-> AS#4 οδηγεί σε συνολικό κόστος της νέας βέλτιστης διαδρομής ίσο με {cost_ttf}.\n')

opt_change = 'AS#2 <-> AS#1' if cost_tto < cost_ttf else 'AS#2 <-> AS#4'
opt_path = path_tto if cost_tto < cost_ttf else path_ttf

print(f'Ως εκ τούτου, η προτιμότερη προσθήκη είναι η {opt_change}, η οποία οδηγεί στο βέλτιστο μονοπάτι (στην περίπτωση βλάβης): {opt_path}')
```

<font color='#486393'>Βάσει των αποτελεσμάτων αυτών, συμπεραίνει κανείς πως πιο συμφέρουσα σε ό,τι έχει να κάνει με την ελαχιστοποίηση του συνολικού κόστους διαδρομής είναι η προσθήκη της ζεύξης AS#2 $\leftrightarrow$ AS#1.</font>

- <font color='#486393'>Το ερώτημα αυτό απαντήθηκε ήδη στα πλαίσια της διερεύνησης που πραγματοποιήθηκε προκειμένου να διαπιστωθεί ποια ζεύξη είναι προτιμότερη σε ό,τι έχει να κάνει με την ελαχιστοποίηση του συνολικού κόστους. Βάσει αποτελεσμάτων, η συνολική επιβάρυνση στην τελική διαδρομή είναι 10 μονάδες, αφού τώρα η βέλτιστη διαδρομή είναι η</font>

<center><font color='#486393'><b>AS#2 $\rightarrow$ AS#1 $\rightarrow$ AS#3 $\rightarrow$ AS#5 $\rightarrow$ AS#7 $\rightarrow$ AS#8 $\rightarrow$ AS#9 $\rightarrow$ AS#11</b>,</font></center>

<font color='#486393'>δηλαδή η ίδια διαδρομή με πριν, μόνο που η μετάβαση AS#2 $\rightarrow$ AS#3 (κόστους 4) έχει αντικατασταθεί από τις μεταβάσεις AS#2 $\rightarrow$ AS#1 $\rightarrow$ AS#3 (κόστους 14). Ο νέος γράφος φαίνεται στο ακόλουθο σχήμα.</font>

```python
DrawGraph(Graph_tto)
```

- <font color='#486393'>Στην περίπτωση αποκατάστασης της βλάβης και την προσθήκη του επιπλέον κόμβου, μαζί με τις σχετικές ζεύξεις, ο γράφος θα έχει τη μορφή που φαίνεται στο ακόλουθο σχήμα.</font>

```python
Nodes_list.append('AS#12')
Edges_list += [("AS#2","AS#1",10),("AS#12","AS#3",2),("AS#12","AS#10",2),("AS#12","AS#11",4)]
Final_graph = CreateGraph(Nodes_list,Edges_list)

DrawGraph(Final_graph)
```

<font color='#486393'>Λύνοντας ξανά το πρόβλημα με τον αλγόριθμο Bellman-Ford, αυτή τη φορά για τον τελικό γράφο, προκύπτουν:</font>

```python
cost_f, path_f, neg_c = bf.bellman_ford(Final_graph, source='AS#2', target='AS#11')
print(f"Το συνολικό κόστος της βέλτιστης διαδρομής του νέου γράφου είναι {cost_f}, ενώ το βέλτιστο μονοπάτι είναι το:")
print(f"{path_f}")
```
