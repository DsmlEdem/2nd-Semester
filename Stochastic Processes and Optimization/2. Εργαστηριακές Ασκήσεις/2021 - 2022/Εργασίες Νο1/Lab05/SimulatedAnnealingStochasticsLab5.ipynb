<h1><b><i>Ο αλγόριθμος Προσομοιωμένης Ανόπτησης</i> (<i>Simulated Annealing</i>)</b></h1>

Στην άσκηση αυτή θα μελετήσετε τον αλγόριθμο της <i>Προσομοιωμένης Ανόπτησης</i> (<i>Simulated Annealing</i>), περισσότερες πληροφορίες για τον οποίο μπορείτε να βρείτε <a href="https://en.wikipedia.org/wiki/Simulated_annealing">εδώ</a>. Θα χρησιμοποιήσετε το <b>έτοιμο</b> πρόγραμμα που δίνεται στη συνέχεια. Το πρόγραμμα αυτό επιλύει το πρόβλημα του <i>Περιπλανώμενου Πωλητή</i> (<i>Travelling Salesman Problem</i>, <i>TSP</i>), περισσότερες πληροφορίες για το οποίο μπορείτε να βρείτε <a href="https://en.wikipedia.org/wiki/Travelling_salesman_problem">εδώ</a>. Αρχικά, το πρόβλημα λύνεται χρησιμοποιώντας τον αλγόριθμο <i>Simulated Annealing</i> και, στη συνέχεια, χρησιμοποιώντας τον αλγόριθμο <a href="https://en.wikipedia.org/wiki/Hill_climbing"><i>Hill Climbing</i></a>. Η επίλυση του προβλήματος γίνεται για <b>150 πόλεις</b>.

```python
#!/usr/bin/env python

""" Traveling salesman problem solved using Simulated Annealing.
http://www.physics.rutgers.edu/~haule/681/src_MC/python_codes/salesman.py
"""
from scipy import *
from pylab import *
from math import floor

import seaborn as sns
sns.set(style = "darkgrid") # καθαρά αισθητικοί λόγοι

mycol = (72/255, 99/255, 147/255)

def Distance(R1, R2):
    return sqrt((R1[0] - R2[0]) ** 2 + (R1[1] - R2[1]) ** 2)

def TotalDistance(city, R):
    dist = 0
    for i in range(len(city) - 1):
        dist += Distance(R[city[i]], R[city[i + 1]])
    dist += Distance(R[city[-1]], R[city[0]])
    return dist
    
def reverse(city, n):
    nct = len(city)
    nn = floor((1 + ((n[1] - n[0]) % nct)) / 2) # half the length of the segment to be reversed
    # the segment is reversed in the following way n[0]<->n[1], n[0]+1<->n[1]-1, n[0]+2<->n[1]-2,...
    # Start at the ends of the segment and swap pairs of cities, moving towards the center.
    for j in range(nn):
        k = (n[0] + j) % nct
        l = (n[1] - j) % nct
        city[k], city[l] = city[l], city[k]
    
def transpt(city, n):
    nct = len(city)
    newcity=[]
    # Segment in the range n[0]...n[1]
    for j in range((n[1] - n[0]) % nct + 1):
        newcity.append(city[(j + n[0]) % nct])
    # is followed by segment n[5]...n[2]
    for j in range((n[2] - n[5]) % nct + 1):
        newcity.append(city[(j + n[5]) % nct])
    # is followed by segment n[3]...n[4]
    for j in range( (n[4] - n[3]) % nct + 1):
        newcity.append(city[(j + n[3]) % nct])
    return newcity

def Plot(city, R, dist):
    # Plot
    Pt = [R[city[i]] for i in range(len(city))]
    Pt += [R[city[0]]]
    Pt = array(Pt)
    title('Total distance='+str(dist))
    plot(Pt[:,0], Pt[:,1], '-o', color=mycol)
    show()

    ###########################################################################################
    ################################# Simulated Annealing Case ################################
    ###########################################################################################

if __name__=='__main__':

    ncity = 150        # Number of cities to visit
    maxTsteps = 100    # Temperature is lowered not more than maxTsteps
    Tstart = 0.2       # Starting temperature - has to be high enough
    fCool = 0.9        # Factor to multiply temperature at each cooling step
    maxSteps = 100 * ncity     # Number of steps at constant temperature
    maxAccepted = 10 * ncity   # Number of accepted steps at constant temperature

    Preverse = 0.5      # How often to choose reverse/transpose trial move

    # Choosing city coordinates
    R=[]  # coordinates of cities are choosen randomly
    for i in range(ncity):
        R.append([rand(), rand()])
    R = array(R)

    # The index table -- the order the cities are visited.
    city = range(ncity)
    city = list(city)
    # Distance of the travel at the beginning
    dist = TotalDistance(city, R)

    # Stores points of a move
    n = zeros(6, dtype = int)
    nct = len(R) # number of cities
    
    T = Tstart # temperature

    Plot(city, R, dist)
    
    for t in range(maxTsteps):  # Over temperature

        accepted = 0
        for i in range(maxSteps): # At each temperature, many Monte Carlo steps
            
            while True: # Will find two random cities sufficiently close by
                # Two cities n[0] and n[1] are choosen at random
                n[0] = int((nct)*rand())     # select one city
                n[1] = int((nct-1)*rand())   # select another city, but not the same
                if (n[1] >= n[0]): n[1] += 1   #
                if (n[1] < n[0]): (n[0],n[1]) = (n[1],n[0]) # swap, because it must be: n[0]<n[1]
                nn = (n[0]+nct -n[1]-1) % nct  # number of cities not on the segment n[0]..n[1]
                if nn>=3: break
        
            # We want to have one index before and one after the two cities
            # The order hence is [n2,n0,n1,n3]
            n[2] = (n[0] - 1) % nct  # index before n0  -- see figure in the lecture notes
            n[3] = (n[1] + 1) % nct  # index after n2   -- see figure in the lecture notes
            
            if Preverse > rand(): 
                # Here we reverse a segment
                # What would be the cost to reverse the path between city[n[0]]-city[n[1]]?
                de = Distance(R[city[n[2]]],R[city[n[1]]]) + Distance(R[city[n[3]]],R[city[n[0]]]) - Distance(R[city[n[2]]],R[city[n[0]]]) - Distance(R[city[n[3]]],R[city[n[1]]])
                
                if de<0 or exp(-de/T)>rand(): # Metropolis
                    accepted += 1
                    dist += de
                    reverse(city, n)
            else:
                # Here we transpose a segment
                nc = (n[1]+1+ int(rand()*(nn-1)))%nct  # Another point outside n[0],n[1] segment. See picture in lecture nodes!
                n[4] = nc
                n[5] = (nc+1) % nct
        
                # Cost to transpose a segment
                de = -Distance(R[city[n[1]]],R[city[n[3]]]) - Distance(R[city[n[0]]],R[city[n[2]]]) - Distance(R[city[n[4]]],R[city[n[5]]])
                de += Distance(R[city[n[0]]],R[city[n[4]]]) + Distance(R[city[n[1]]],R[city[n[5]]]) + Distance(R[city[n[2]]],R[city[n[3]]])
                
                if de<0 or exp(-de/T)>rand(): # Metropolis
                    accepted += 1
                    dist += de
                    city = transpt(city, n)
                    
            if accepted > maxAccepted: break

        # Plot
        Plot(city, R, dist)
            
        print("Step=%d, T=%10.5f , distance= %10.5f , accepted steps= %d" %(t, T, dist, accepted))
        T *= fCool             # The system is cooled down
        if accepted == 0: break  # If the path does not want to change any more, we can stop
        
    ###########################################################################################
    ############################# Repeating for Hill Climbing Case ############################
    ###########################################################################################
    print("\n"+64*"*"+"\n")
    print("Repeating for Hill Climbing Case\n")
    
    # The index table -- the order the cities are visited.
    city = range(ncity)
    city = list(city)
    # Distance of the travel at the beginning
    dist = TotalDistance(city, R)

    # Stores points of a move
    n = zeros(6, dtype=int)
    nct = len(R) # number of cities
    
    T = Tstart # temperature

    Plot(city, R, dist)
    
    for t in range(maxTsteps):  # Over temperature

        accepted = 0
        for i in range(maxSteps): # At each temperature, many Monte Carlo steps
            
            while True: # Will find two random cities sufficiently close by
                # Two cities n[0] and n[1] are choosen at random
                n[0] = int((nct)*rand())     # select one city
                n[1] = int((nct-1)*rand())   # select another city, but not the same
                if (n[1] >= n[0]): n[1] += 1   #
                if (n[1] < n[0]): (n[0],n[1]) = (n[1],n[0]) # swap, because it must be: n[0]<n[1]
                nn = (n[0]+nct -n[1]-1) % nct  # number of cities not on the segment n[0]..n[1]
                if nn>=3: break
        
            # We want to have one index before and one after the two cities
            # The order hence is [n2,n0,n1,n3]
            n[2] = (n[0]-1) % nct  # index before n0  -- see figure in the lecture notes
            n[3] = (n[1]+1) % nct  # index after n2   -- see figure in the lecture notes
            
            if Preverse > rand(): 
                # Here we reverse a segment
                # What would be the cost to reverse the path between city[n[0]]-city[n[1]]?
                de = Distance(R[city[n[2]]],R[city[n[1]]]) + Distance(R[city[n[3]]],R[city[n[0]]]) - Distance(R[city[n[2]]],R[city[n[0]]]) - Distance(R[city[n[3]]],R[city[n[1]]])
                
                if de<0: # Hill Climbing
                    accepted += 1
                    dist += de
                    reverse(city, n)
            else:
                # Here we transpose a segment
                nc = (n[1]+1+ int(rand()*(nn-1)))%nct  # Another point outside n[0],n[1] segment. See picture in lecture nodes!
                n[4] = nc
                n[5] = (nc+1) % nct
        
                # Cost to transpose a segment
                de = -Distance(R[city[n[1]]],R[city[n[3]]]) - Distance(R[city[n[0]]],R[city[n[2]]]) - Distance(R[city[n[4]]],R[city[n[5]]])
                de += Distance(R[city[n[0]]],R[city[n[4]]]) + Distance(R[city[n[1]]],R[city[n[5]]]) + Distance(R[city[n[2]]],R[city[n[3]]])
                
                if de<0: # Hill Climbing
                    accepted += 1
                    dist += de
                    city = transpt(city, n)
                    
            if accepted > maxAccepted: break

        # Plot
        Plot(city, R, dist)
            
        print("Step=%d, T=%10.5f , distance= %10.5f , accepted steps= %d" %(t, T, dist, accepted))
        T *= fCool             # The system is cooled down
        if accepted == 0: break  # If the path does not want to change any more, we can stop
```

- <font color='#486393'>Δεδομένων των παθογενειών του αλγορίθμου Hill Climbing, μια από τις προτεινόμενες τροποποιήσεις του μέσω στοχαστικών μεθόδων είναι αυτή της προσομοιωμένης ανόπτησης (simulated annealing). Η έννοια της ανόπτυσης είναι δανεισμένη από τη Φυσική Στερεάς Κατάστασης, όπου η δομή και οι ιδιότητες ορισμένων υλικών μεταβάλλονται μέσω επανάληψης της διαδικασίας</font>

<font color='#486393'><center>θέρμανσης σε υψηλή θερμοκρασία → σταθεροποίηση της θερμοκρασίας σε μια υψηλή τιμή → ψύξη του υλικού</center></font>

<font color='#486393'>αρκετές φορές. Στα πλαίσια της στοχαστικής βελτιστοποίησης, η προσομοιωμένη ανόπτηση θυμίζει σε μεγάλο βαθμό τον αλγόριθμο Metropolis-Hastings που αναλύθηκε στο προηγούμενο εργαστήριο.</font>

<font color='#486393'>Συγκεκριμένα για το πρόβλημα του περιπλανώμενου πωλητή, μπορούμε στο μήκος κάθε προτεινόμενης διαδρομής $M$ να αντιστοιχίσουμε μια «ενέργεια» $E\left(M\right)$ έτσι, ώστε η ενέργεια να ελαχιστοποιείται όταν ελαχιστοποιείται και το μήκος της αντίστοιχης διαδρομής (θα μπορούσαμε να θεωρήσουμε και $E\left(M\right) = M$, όμως επιλέγουμε να παρουσιάσουμε μια πιο γενική προσέγγιση). Κατά το διαταρακτικό στάδιο της επαναληπτικής διαδικασίας που περιλαμβάνει ο αλγόριθμος Hill Climbing, δεδομένης μιας λύσης με μήκος $M$ διερευνώνται πιθανές διαταρακτικές λύσεις με μήκος $M^\prime$. Εκεί, εάν $E\left(M\right) < E\left(M^\prime\right)$, τότε η προτεινόμενη διαταραχή απορρίπτεται με βεβαιότητα, ενώ αντίθετα γίνεται αποδεκτή στην περίπτωση όπου $E\left(M\right) > E\left(M^\prime\right)$. Αντιθέτως, στην περίπτωση της προσομοιωμένης ανόπτυσης η επιλογή ή η απόρριψη μιας διαταραχής στην υπάρχουσα λύση γίνεται πιθανολογικά: αφού πρώτα υπολογιστεί η ποσότητα</font>

<font color='#486393'>$$ p = \exp\left(-\frac{E\left(M^\prime\right) - E\left(M\right)}{T}\right),$$</font>

<font color='#486393'>γεννάται στο διάστημα $[0,1]$ ένας ψευδοτυχαίος αριθμός $\chi$. Η διαταραχή γίνεται αποδεκτή εάν $p > \chi$, διαφορετικά απορρίπτεται. Στην άνω σχέση, η ποσότητα $T$ αντιστοιχεί σε μια θερμοκρασία η οποία ξεκινά από μια πεπερασμένη μη μηδενική τιμή και μειώνεται με σταθερό ρυθμό από επανάληψη σε επανάληψη του αλγορίθμου, μέχρι να φτάσει σε ένα κάτω κατώφλι (εξ ου και η έμπνευση για το όνομα της μεθόδου). Με τον τρόπο αυτό, κατά τα πρώτα βήματα του αλγορίθμου υπάρχει μια **σχετικά** υψηλή πιθανότητα να επιλεγεί μια λύση η οποία να οδηγεί σε μεγαλύτερα μήκη διαδρομών (διευκολύνοντας έτσι την «απόδραση» από κακές αρχικοποιήσεις), ενώ όσο η θερμοκρασία μειώνεται, μειώνεται και η πιθανότητα αυτή.</font>

<font color='#486393'>Ο τρόπος με τον οποίο ο αλγόριθμος αυτός δίνει καλύτερες λύσεις σε σχέση με τον αλγόριθμο Hill Climbing για το πρόβλημα του περιπλανώμενου πωλητή είναι πως είναι διατεθειμένος να «θυσιάσει» μια λύση η οποία προσεγγίζει με υψηλή ακρίβεια ένα τοπικό μέγιστο, προκειμένου να αναζητήσει μια λύση η οποία να προσεγγίζει το ολικό μέγιστο, ακόμη κι αν η ακρίβεια δεν είναι τόσο καλή. Αυτό, διότι η λιγότερο ακριβής προσέγγιση του ολικού μεγίστου μπορεί κάλλιστα να αντιστοιχεί σε μια διαδρομή μικρότερου συνολικού μήκους σε σχέση με αυτήν που προκύπτει από το εκάστοτε τοπικό μέγιστο. Αξίζει να σημειώσουμε πως οι δύο αλγόριθμοι είναι ισοδύναμοι στο όριο της μηδενικής θερμοκρασίας $T \to 0$, όπου στην ουσία η συνάρτηση Boltzmann ($p$) αναγάγεται στη βηματική συνάρτηση.</font>

Ο αλγόριθμος προσομοιωμένης ανόπτυσης επιτυγχάνει ένα καλύτερο τελικό αποτέλεσμα για το ελάχιστο μήκος διαδρομής, ίσο με $d_\text{sa} = 9.71$, σε αντίθεση με την τιμή $d_\text{hc} = 10.59$ που προκύπτει από τον αλγόριθμο Hill Climbing. Από την άλλη, ο Hill Climbing συγκλίνει πολύ γρηγορότερα, σε μόλις 9 επαναλήψεις έναντι των 53 επαναλήψεων του αλγορίθμου προσομοιωμένης ανόπτυσης. Το αποτέλεσμα αυτό είναι αναμενόμενο, αφού ο Hill Climbing αναζητά το τοπικό μέγιστο που βρίσκεται πιο κοντά στη θέση που αντιστοιχεί στην αρχικοποίησή του στο χώρο των λύσεων, ενώ ο αλγόριθμος προσομοιωμένης ανόπτυσης αναζητά το ολικό μέγιστο στο χώρο αυτό. Το κατά πόσο η ελαχιστοποίηση του μήκους της διαδρομής υπερτερεί έναντι της ταχύτητας σύγκλισης έχει φυσικά να κάνει με τις απαιτήσεις του καθενός, καθώς και με το πόσο μεγάλες είναι οι αποκλίσεις σε κάθε ένα από αυτά τα δύο χαρακτηριστικά από αλγόριθμο σε αλγόριθμο. Αυτός είναι και ο λόγος για τον οποίο δε μπορεί κανείς να μιλήσει με απόλυτους όρους για «καλύτερο» αλγόριθμο ανάμεσα στους δύο.</font>

- <font color='#486393'>Με εξαίρεση τη μετάβαση από την πρώτη επανάληψη των αλγορίθμων στη δεύτερη, και οι δύο αλγόριθμοι χαρακτηρίζονται από σχετικά ήπιες μεταβολές από λύση σε λύση, γεγονός το οποίο οφείλεται στην αναζήτηση **μικρών** διαταραχών η οποία συζητήθηκε προηγουμένως. Η παρατήρηση αυτή απαντά και στο δεύτερο ερώτημα του συγκεκριμένου ζητήματος (το οποίο, βέβαια, αναλύθηκε εκτενώς και στα προηγούμενα): ο αλγόριθμος που από βήμα σε βήμα επιλέγει μόνο «βελτιωμένες» διαδρομές είναι ο Hill Climbing, ενώ ο Simulated Annealing επιτρέπει και μεταβάσεις σε λύσεις με μεγαλύτερα μήκη διαδρομής.
