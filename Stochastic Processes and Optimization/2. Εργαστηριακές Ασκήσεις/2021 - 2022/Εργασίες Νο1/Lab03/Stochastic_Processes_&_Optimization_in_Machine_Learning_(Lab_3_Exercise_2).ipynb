### Απαντήσεις

- <font color='#486393'>Όπως έχει αναλυθεί και προηγουμένως, το άθροισμα κάθε γραμμής του πίνακα μετάβασης θα πρέπει να ισούται με τη μονάδα. Ως εκ τούτου, η μήτρα $P$ φαίνεται ακολούθως, με τα συμπληρωμένα κενά να αναγράφονται σε bold:</font>
    
<font color='#486393'>$$ P = \begin{pmatrix}
0.6 & \mathbf{0.2} & 0 & 0 & 0 & 0.2 \\
0.2 & \mathbf{0.6} & 0.2 & 0 & 0 & 0 \\
0 & 0.2 & 0.6 & 0.2 & \mathbf{0} & 0 \\
0 & 0 & 0.2 & \mathbf{0.6} & 0.2 & 0 \\
0 & \mathbf{0} & \mathbf{0} & 0.2 & 0.6 & 0.2 \\
0.2 & 0 & 0 & 0 & \mathbf{0.2} & 0.6
\end{pmatrix} $$ </font>

- <font color='#486393'>Όπως φαίνεται από τη μορφή του πίνακα μετάβασης, η αλυσίδα περιγράφεται από τις καταστάσεις $1$, $2$, $\dots$, $6$, καθεμιά εκ των οποίων έχει τη δυνατότητα μετάβασης στην επόμενη ή την προηγούμενη (αριθμητικά) με πιθανότητα 0.2 και τη δυνατότητα παραμονής με πιθανότητα 0.6. Η παρατήρηση αυτή συνοψίζεται στο ακόλουθο σχήμα, όπου απεικονίζεται η αλυσίδα.</font>

<font color='#486393'>Υπενθυμίζεται στο σημείο αυτό πως στην ίδια κλάση επικοινωνίας ανήκουν όλες οι καταστάσεις οι οποίες ανά δύο συνδέονται μέσω κάποιας ακολουθίας μεταβάσεων με μη μηδενική πιθανότητα. Δεδομένου αυτού, γίνεται εμφανές πως στην προκείμενη περίπτωση όλες οι καταστάσεις ανήκουν στην ίδια κλάση επικοινωνίας, την $\{1,2,3,4,5,6\}$, το οποίο υποδεικνύει πως η υπό μελέτη αλυσίδα Markov είναι <i>μη αναγωγίσιμη</i> (στις σημειώσεις του Μ. Λουλάκη τέτοιες αλυσίδες αναφέρονται ως <i>μη υποβιβάσιμες</i>, αν και αυτός ο όρος είναι λιγότερο συχνός στη βιβλιογραφία). Ο λόγος για αυτό είναι πως κάθε κλειστή πεπερασμένη κλάση είναι και επαναληπτική.</font>

<font color='#486393'>Δεδομένου πως η αλυσίδα Markov που παρουσιάζεται εδώ είναι μη αναγωγίσιμη, σε συνδυασμό με το γεγονός πως είναι πεπερασμένη (δηλαδή δομείται από πεπερασμένο πλήθος καταστάσεων), οδηγείται κανείς στο συμπέρασμα πως για την αλυσίδα αυτή υπάρχει αναλλοίωτη κατανομή $\pi = \left(\pi_1, \dots, \pi_6\right)$, η οποία ικανοποιεί τις σχέσεις $\pi = \pi P$ και $\sum_{i=1}^{6}{\pi_i} = 1$. Επιπροσθέτως, από την κατανομή αυτή δε θα αποκλείεται οποιαδήποτε κατάσταση: θα ισχύει $\pi_i \neq 0$, για κάθε $i = 1, \dots, 6$, ή, ισοδύναμα, αφότου παρέλθει το μεταβατικό φαινόμενο, κάθε κατάσταση θα μπορεί να εμφανιστεί διαδοχικά σε πεπερασμένο πλήθος βημάτων $E_i = \frac{1}{\pi_i}$. Τέλος, υπάρχει μια ακόμη ιδιότητα που δεν αφορά τις μη αναγωγίσιμες πεπερασμένες αλυσίδες Markov εν γένει, αλλά την αλυσίδα Markov που παρουσιάζεται εδώ: δεδομένου πως το άθροισμα όχι μόνο των γραμμών, αλλά και των στηλών του πίνακα μετάβασης ισούται με τη μονάδα, ο πίνακας $P$ καλείται <i>διπλά στοχαστικός</i>. Σε πρακτικό επίπεδο, αυτό σημαίνει πως η αναλλοίωτη κατανομή είναι εκ προοιμίου γνωστή και ίση με $\pi = \left(\frac{1}{6},\frac{1}{6},\frac{1}{6},\frac{1}{6},\frac{1}{6},\frac{1}{6}\right)$.</font>

- <font color='#486393'>Λαμβάνοντας υπ' όψιν τις παραπάνω παρατηρήσεις, προχωρούμε στην εκτίμηση της αναλλοίωτης κατανομής μέσω προσομοίωσης, έχοντας όμως το πλεονέκτημα να γνωρίζουμε εκ των προτέρων το επιθυμητό αποτέλεσμα. Ο σχετικός κώδικας (ο οποίος ταυτίζεται σχεδόν με τον αρχικό κώδικα, όμως έχει υποστεί ορισμένες τροποποιήσεις για αισθητικούς και μη λόγους) φαίνεται ακολούθως.</font>

```python
from __future__ import division
from simple_markov_chain_lib import markov_chain
import numpy as np
import matplotlib.pyplot as plt

mycol = (72/255, 99/255, 147/255)

def defineMarkovTable(): 
    markov_table = {
        0: {0: 0.6, 1: 0.2, 5: 0.2},
        1: {0: 0.2, 1: 0.6, 2: 0.2},
        2: {1: 0.2, 2: 0.6, 3: 0.2},
        3: {2: 0.2, 3: 0.6, 4: 0.2},
        4: {3: 0.2, 4: 0.6, 5: 0.2},
        5: {0: 0.2, 4: 0.2, 5: 0.6}
        }

    return markov_table

def defineInitDistribution():
    init_dist = {0: 1.}

    return init_dist

def calculateFrequencyTable(mc,markov_table,init_dist,total_runs):

    visits = list()
    visits = [0] * len(markov_table)

    for index in range(total_runs):
        mc.move()
        visits[mc.running_state] += 1

    frequency = [v/total_runs for v in visits]

    return frequency

def printFrequencies(frequency, markov_table):
    error = 0.0
    for state in range(0, len(markov_table)) :
        print(f"Συχνότητα εμφάνισης της κατάστασης {state+1}: {frequency[state]:.4f}.")
        error += abs(1/6-frequency[state])
    print(f"\nΗ μέση απόκλιση από τη θεωρητική τιμή ισούται με: {(1.0/6.0)*error:.4f}.\n")

    return None

def plotFrequencies(frequency,total_runs):
    objects = ('1', '2', '3', '4', '5', '6')
    y_pos = np.arange(len(objects))

    plt.bar(y_pos, frequency, align='center', alpha=0.5, color=mycol)
    plt.xticks(y_pos, objects)
    plt.xlabel('Κατάσταση')
    plt.ylabel('Συχνότητα Εμφάνισης')
    plt.title(f'Ιστόγραμμα συχνοτήτων για {total_runs} επαναλήψεις')
    plt.show()

    return None

def run_markov(total_runs):
    markov_table = defineMarkovTable()
    init_dist = defineInitDistribution()
    mc = markov_chain(markov_table, init_dist)
    mc.start()

    frequency = calculateFrequencyTable(mc,markov_table, init_dist, total_runs)
    plotFrequencies(frequency, total_runs)
    printFrequencies(frequency, markov_table)
    
    return None

import seaborn as sns
sns.set(style = "darkgrid") # καθαρά αισθητικοί λόγοι
```

<font color='#486393'>Χρησιμοποιώντας τη συνάρτηση `run_markov`, πραγματοποιείται εκτέλεση της διαδικασίας για (α) 500 (β) 1000 (γ) 10000 (δ) 100000 (ε) 1000000 (στ) 10000000 και (ζ) 100000000 μεταβάσεις στην αλυσίδα. Σε κάθε κελί φαίνεται από μία εκτέλεση, μαζί με τα αποτελέσματά της καθώς και το συνολικό χρόνο που απαιτήθηκε. Σημειώνεται πως οι αναγραφόμενοι χρόνοι δεν έχουν νόημα ως απόλυτα μεγέθη, αφού εξαρτώνται από τον επεξεργαστή του συστήματος όπου εκτελέστηκε ο κώδικας. Αντίθετα, φυσική σημασία έχουν οι σχετικές τιμές τους.</font>
