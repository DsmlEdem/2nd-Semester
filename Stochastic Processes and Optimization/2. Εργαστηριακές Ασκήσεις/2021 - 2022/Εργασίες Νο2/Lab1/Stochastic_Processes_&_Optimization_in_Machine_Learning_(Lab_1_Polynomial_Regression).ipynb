```python
#!pip install numpy
#!pip install matplotlib
#!pip install pandas
#!pip install sklearn

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

Στη συνέχεια, θα φορτώσετε τα δεδομένα εκπαίδευσης που παρέχονται από το αρχείο
 <b><i>data2.csv</b></i>.

```python
datas = pd.read_csv('data2.csv')
print(datas)
```

Τώρα, θα δημιουργήσετε τις μεταβλητές εισόδου και εξόδου για τους αλγορίθμους μηχανικής μάθησης από τα δεδομένα εκπαίδευσης που παρέχονται.

```python
X = datas.iloc[:, 1:2].values
y = datas.iloc[:, 2].values

print("Input")

print(X)

print("Labels")

print(y)
```

Στη συνέχεια, θα εκπαιδεύσετε το μοντέλο <b><i>linear regression</i></b>, χρησιμοποιώντας τα παραπάνω δεδομένα εκπαίδευσης.

```python
lin = LinearRegression()
lin.fit(X, y)
```

<font color='#486393'>Τα σχόλια στα βήματα της παρούσας αναφοράς γράφονται σε αυτό το χρώμα, προκειμένου να διαφοροποιούνται από την ίδια την αναφορά.</font>

<h3><b><i>Ερώτημα 1</b></i></h3>

<p align="justify">Να εκτελέσετε τα παρακάτω τμήματα κώδικα για τις τιμές της παραμέτρου <b><i>degree</b></i> {<i>2, 3, 4, 5, 8, 10, 12, 15</i>} και να καταγράψετε τα διαγράμματα που παρέχονται ως έξοδοι. Τι αλλαγές παρατηρείτε στη μορφή της γραμμής παλινδρόμησης για το μοντέλο <b><i>polynomial regression</i></b>, καθώς αυξάνεται η τιμή της παραμέτρου <b><i>degree</i></b>; Τι παρατηρείτε για τον απαιτούμενο χρόνο εκπαίδευσης του μοντέλου <b><i>polynomial regression</i></b>, καθώς αυξάνεται η τιμή της παραμέτρου <b><i>degree</i></b>;</p>

<br>

**<h4> <b><i>degree</b></i> = 2 </h4>**

```python
%%time
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
poly.fit(X_poly, y)
lin2 = LinearRegression()
lin2.fit(X_poly, y)

plt.scatter(X, y, color = 'blue', label = 'Training Examples')

plt.plot(X, lin.predict(X), color = 'red', label = 'Linear Regression Line')
plt.plot(X, lin2.predict(poly.fit_transform(X)), color = 'green', label = 'Polynomial Regression Line')
plt.legend()
plt.show()
```

<font color='#486393'>Δεδομένου βαθμού </font>$d$<font color='#486393'>, η προσαρμογή γίνεται σε κάθε περίπτωση σε ένα πολυώνυμο της μορφής</font>

$$p\left(x\right) = \alpha_0 + \sum_{n=1}^{d}{\alpha_n x^n}.$$

<font color='#486393'>Προφανώς, όσο υψηλότερη είναι η τιμή του </font>$d$<font color='#486393'>, με τόσο μεγαλύτερη ακρίβεια μπορεί η υπολογισμένη καμπύλη να προσαρμοστεί στα δεδομένα εκπαίδευσης.</font>

<font color='#486393'>Φυσικά, η υπερπροσαρμογή στα δεδομένα εκπαίδευσης δεν είναι επιθυμητή, αφού μειώνει σημαντικά την ικανότητα γενίκευσης του μοντέλου: παρότι το εκπαιδευμένο μοντέλο έχει καλή προβλεπτική ικανότητα για τα δεδομένα επάνω στα οποία έχει εκπαιδευτεί, όταν καλείται να εφαρμοστεί σε νέα δεδομένα εμφανίζει υψηλά σφάλματα. Επιπλέον, αυξάνει το βαθμό επιρροής που έχουν στο μοντέλο πιθανοί outliers που υπάρχουν στο δείγμα.</font>

<font color='#486393'>Στόχος στην πολυωνυμική παλινδρόμηση είναι η εύρεση ενός βαθμού </font>$d$ <font color='#486393'>τέτοιου, ώστε να επιτυγχάνεται μια «χρυσή τομή» ανάμεσα στην πολυπλοκότητα του μοντέλου, δηλαδή την ικανότητά του να περιγράψει κατάλληλα τις συσχετίσεις ανάμεσα στα υπό μελέτη χαρακτηριστικά, και στη δυνατότητα γενίκευσής του. Αυτό είναι το γνωστό <i>bias-variance trade-off</i>. Σε κάθε περίπτωση, η εύρεση του κατάλληλου αυτού βαθμού είναι εφικτή με σχετική ασφάλεια μόνο όταν υπάρχει ένα δειγματικό σύνολο σχετικά υψηλής πληθικότητας, γεγονός που δεν ισχύει στην προκείμενη περίπτωση (μόλις 15 σημεία). Παρ' όλα αυτά, οπτικά φαίνεται τα πολυώνυμα βαθμού 3 και 4 να προσαρμόζονται σχετικά ικανοποιητικά στα δεδομένα, χωρίς όμως η προσαρμογή αυτή να γίνεται σε υπερβολικό βαθμό.</font>

<font color='#486393'>Σε υπολογιστικό επίπεδο, αξίζει να αναφερθεί πως η αύξηση του βαθμού του πολυωνύμου οδηγεί σε μη γραμμική αύξηση του χρόνου εκπαίδευσης του μοντέλου, όπως φαίνεται από το output των παραπάνω κελιών, στα οποία έχει προστεθεί η εντολή </font>`%%time` <font color='#486393'>. Ενώ για </font>$d = 12$<font color='#486393'> το wall time είναι </font>$\mathcal{O}\left(10^2\right)$<font color='#486393'> ms, για </font>$d = 15$<font color='#486393'> είναι </font>$\mathcal{O}\left(10^4\right)$ <font color='#486393'>ms, δηλαδή 2 τάξεις μεγέθους υψηλότερο. Σημειώνεται πως ο χρόνος αυτός υπολογίζεται χρησιμοποιώντας επεξεργαστή Intel Core i7-10700KF.

<h3><b><i>Ερώτημα 2</b></i></h3>

Σας παρέχεται, επίσης, το αρχείο <b><i>data2b.csv</i></b>, το οποίο είναι το αρχείο <b><i>data2.csv</i></b> με μία παραπάνω εγγραφή, την <b><i>(9, 38, 0.3)</i></b>. Να εκτελέσετε το παρακάτω πρόγραμμα για την τιμή της παραμέτρου <b><i>degree = 15</i></b>. Πόσο μεγάλη είναι η αλλαγή που παρατηρείτε στη μορφή της γραμμής παλινδρόμησης σε σχέση με την αντίστοιχη περίπτωση του ερωτήματος 1; Τι συμπέρασμα μπορείτε να βγάλετε για τον αλγόριθμο <b><i>polynomial regression</i></b>;

```python
datas = pd.read_csv('data2b.csv')
print(datas)

X = datas.iloc[:, 1:2].values
y = datas.iloc[:, 2].values

lin = LinearRegression()
lin.fit(X, y)

poly = PolynomialFeatures(degree=15)
X_poly = poly.fit_transform(X)
poly.fit(X_poly, y)
lin2 = LinearRegression()
lin2.fit(X_poly, y)

plt.scatter(X, y, color = 'blue', label = 'Training Examples')

plt.plot(X, lin.predict(X), color = 'red', label = 'Linear Regression Line')
plt.plot(X, lin2.predict(poly.fit_transform(X)), color = 'green', label = 'Polynomial Regression Line')
plt.legend()
plt.show()
```

<font color='#486393'>Γίνεται εμφανές πως η προσθήκη του νέου σημείου μεταβάλλει σημαντικά την εικόνα της προσαρμοσμένης καμπύλης στην περίπτωση </font>$d = 15$<font color='#486393'>. Το γεγονός αυτό οφείλεται στην προαναφερθείσα υπερπροσαρμογή του μοντέλου στα δεδομένα. Η υπερπροσαρμογή αυτή εντείνει σε πολύ υψηλό βαθμό την επιρροή του νέου σημείου στο μοντέλο, γεγονός το οποίο έχει πρόσθετη βαρύτητα στη συγκεκριμένη περίπτωση, μιας και το σημείο αυτό φαίνεται να αποτελεί outlier, παρατηρώντας τα υπόλοιπα δεδομένα του δειγματικού συνόλου. Έτσι, η καμπύλη παραμορφώνεται σημαντικά, αυξάνοντας την απόστασή της από 4 σημεία εκπαίδευσης, στην προσπάθειά της να «πλησιάσει» το νέο αυτό σημείο, προκειμένου να μειώσει το μέσο τετραγωνικό σφάλμα.</font>
