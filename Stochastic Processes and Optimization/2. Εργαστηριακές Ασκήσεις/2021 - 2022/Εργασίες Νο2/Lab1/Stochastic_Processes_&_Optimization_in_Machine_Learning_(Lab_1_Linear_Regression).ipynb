Σημειώνεται πως οι απαντήσεις στα ερωτήματα γράφονται σε αυτό το χρώμα, προκειμένου να διαφοροποιούνται από την εκφώνηση της άσκησης.</font>

## Μέρος Α: Κατασκευή του Pipeline

```python
#!pip install numpy
#!pip install matplotlib
#!pip install pandas

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.utils import shuffle
from sklearn.metrics import mean_squared_error as MSE

import seaborn as sns
sns.set(style = "darkgrid") # καθαρά αισθητικοί λόγοι

from sklearn import linear_model as LM # για μια τελική σύγκριση

def pipeline(data,split_ratio=0.2,lr=0.03,epochs=10,apply_minmax=True):
    """
    Args:
        - data (pandas dataframe):
            Το σύνολο των υπό μελέτη δεδομένων.
        - split_ratio (float):
            Ένας αριθμός στο διάστημα [0,1] που καθορίζει το ποσοστό
            του συνόλου των δεδομένων το οποίο θα χρησιμοποιηθεί για
            το evaluation. Προφανώς 1-split_ratio είναι το ποσοστό που
            θα χρησιμοποιηθεί για την εκπαίδευση.
        - lr (float):
            Το learning rate για την εκμάθηση του μοντέλου.
        - epochs (int):
            Το πλήθος εποχών εκπαίδευσης του μοντέλου.
        - apply_minmax (boolean):
            Παράμετρος που ελέγχει εάν εφαρμόζεται Min/Max Scaling ή όχι.
    """
    
    X, y = np.asarray(data.iloc[:,0]), np.asarray(data.iloc[:,1])
    if apply_minmax:
        print(f"Τα δεδομένα φορτώθηκαν. Η ανεξάρτητη μεταβλητή είναι η x = {data.columns[0]}, ενώ η εξαρτημένη η y = {data.columns[1]}.\n")
    
        correl = data.corr().iloc[0,1]
        print(f"Η συσχέτιση ανάμεσα στην ανεξάρτητη και την εξαρτημένη μεταβλητή είναι ίση με {correl*100:.2f}%.\n")
    
        mycol = (72/255, 99/255, 147/255)
    
        print("Ακολούθως απεικονίζεται το διάγραμμα διασποράς των δύο μεταβλητών.\n")

        fig, ax = plt.subplots()
        ax.scatter(X, y, color=mycol)
        ax.set_xlabel(data.columns[0])
        ax.set_ylabel(data.columns[1])
        ax.set_title('Διάγραμμα διασποράς των δεδομένων')
        plt.show()
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = split_ratio, random_state = 420)
    if apply_minmax:
        print(f"\nΟ διαχωρισμός των δεδομένων σε train/test έγινε σε ποσοστό {split_ratio*100:.2f}% για τα test δεδομένα.\n")
    
        mmscaler_x, mmscaler_y = MinMaxScaler(), MinMaxScaler()
        X_train_sc = mmscaler_x.fit_transform(X_train.reshape(-1, 1))
        X_test_sc = mmscaler_x.transform(X_test.reshape(-1, 1))

        y_train_sc = mmscaler_y.fit_transform(y_train.reshape(-1, 1))
        y_test_sc = mmscaler_y.transform(y_test.reshape(-1, 1))
        
        X_train, y_train = X_train_sc.reshape(-1,), y_train_sc.reshape(-1,)
        X_test, y_test = X_test_sc.reshape(-1,), y_test_sc.reshape(-1,)
        
        print(f"Πραγματοποιήθηκε το κατάλληλο Min/Max Scaling, οπότε ξεκινά η εκπαίδευση του μοντέλου.\n")
    
    # Applying the SGD
    w_0, w_1 = 0.0, 0.0
    mse_list = []
    for epoch in range(epochs):
        X_shuff, y_shuff = shuffle(X_train, y_train, random_state=420) # stochasticity
        for x, y in zip(X_shuff, y_shuff):
            diff = y - (w_0 + w_1*x)
            w_0 -= lr*(-2.0*diff)
            w_1 -= lr*(-2.0*x*diff)
        mse_list.append(MSE(y_train,w_0+w_1*X_train))
            
    MSE_test = MSE(y_test,w_0+w_1*X_test)
    print(f"Μετά από εκπαίδευση για {epochs} εποχές με ρυθμό εκμάθησης {lr:.2f}, το μοντέλο αντιστοιχεί στην ευθεία:\n")
    print(f"y = {w_0:.3f} + {w_1:.3f}x.\n")
    
    print("Στα ακόλουθα σχήματα απεικονίζεται το μέσο σφάλμα εκπαίδευσης ανά εποχή, καθώς και η προσαρμογή της ευθείας στα test δεδομένα.\n")
    
    mycomplcol = (129/255, 143/255, 163/255)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,6))
    
    ax1.plot(range(1,epochs+1), mse_list, '+', color=mycol)
    ax1.set_xlabel('Εποχή')
    ax1.set_xticks(range(1,epochs+1,3))
    ax1.set_ylabel('Μέσο Τετραγωνικό Σφάλμα')
    ax1.set_title('Μέσο Τετραγωνικό Σφάλμα (MSE) ανά Εποχή εκπαίδευσης')
    
    ax2.scatter(X_test, y_test, color=mycol)
    ax2.plot(X_test, w_0+w_1*X_test, color=mycomplcol)
    ax2.set_xlabel(str(data.columns[0])+' normalized')
    ax2.set_ylabel(str(data.columns[1])+' normalized')
    ax2.set_title('Διάγραμμα Διασποράς & Προσαρμοσμένη Ευθεία')
    
    plt.show()
    
    print(f"Το μέσο τετραγωνικό σφάλμα για τα test δεδομένα προκύπτει ίσο με {MSE_test:.4f}.\n")
    
    reg = LM.LinearRegression()
    reg.fit(X_train.reshape((-1, 1)),y_train.reshape((-1, 1)))
    
    print(f"Ως αναφορά, η ευθεία που προκύπτει για τα δεδομένα εκπαίδευσης χρησιμοποιώντας τη μέθοδο ελαχίστων τετραγώνων είναι η:\n")
    
    reg_w_0, reg_w_1 = reg.intercept_[0], reg.coef_[0][0]
    
    print(f"y = {reg_w_0:.3f} + {reg_w_1:.3f}x.\n")
    
    print(f"Το αντίστοιχο μέσο τετραγωνικό σφάλμα για τα test δεδομένα προκύπτει ίσο με {MSE(y_test,reg_w_0+reg_w_1*X_test):.4f}.\n")
```

<font color='#486393'>Στο παραπάνω pipeline, αρχικά τα δεδομένα διαχωρίζονται σε χαρακτηριστικά (X) και targets (y). Κατόπιν, υπολογίζεται η συσχέτιση μεταξύ X και y, προκειμένου να υπάρξει μια πρώτη εικόνα ως προς το κατά πόσο η εφαρμογή παλινδρόμησης έχει νόημα. Μια δεύτερη εικόνα σχετικά με αυτό αποκτάται με την απεικόνιση του διαγράμματος διασποράς ανάμεσα σε X και y.</font>

<font color='#486393'>Στη συνέχεια, τα δεδομένα χωρίζονται σε δύο υποσύνολα: ένα υποσύνολο εκπαίδευσης (train) σε ποσοστό</font> `(1-split_ratio)*100%` <font color='#486393'>και ένα υποσύνολο αξιολόγησης (test) σε ποσοστό</font> `split_ratio*100%` <font color='#486393'>. Στα train δεδομένα εφαρμόζεται η κανονικοποίηση Min/Max εκπαιδεύοντας έναν</font> `MinMaxScaler` <font color='#486393'>της sklearn και κατόπιν ο εκπαιδευμένος scaler εφαρμόζεται και στα test δεδομένα. Είναι σημαντικό εδώ να τονιστούν τα ακόλουθα:</font>

- <font color='#486393'>Σε μια ρεαλιστική εφαρμογή, τα δεδομένα αξιολόγησης δεν είναι διαθέσιμα παρά μόνο αφότου το μοντέλο έχει ήδη περάσει στο στάδιο του deployment. Έτσι, οποιοσδήποτε μετασχηματισμός δε μπορεί να πραγματοποιηθεί στο δειγματικό σύνολο, αλλά μόνο στο υποσύνολο των train δεδομένων, διότι διαφορετικά το μοντέλο λαμβάνει πληροφορία για δεδομένα τα οποία ρεαλιστικά δεν έχει δει ποτέ.</font>

- <font color='#486393'>Οι μετασχηματισμοί που θα εφαρμοστούν στα δεδομένα εκπαίδευσης πρέπει να εφαρμοστούν αυτούσιοι και στα δεδομένα αξιολόγησης, πριν την αξιολόγηση του μοντέλου, αφού αποτελούν μέρος του pipeline που ορίζει το μοντέλο. Αυτός είναι και ο λόγος για τον οποίο ο MinMaxScaler εκπαιδεύεται στα δεδομένα εκπαίδευσης και εφαρμόζεται αυτούσιος και στα δεδομένα αξιολόγησης. Εάν, για παράδειγμα, τα δεδομένα αξιολόγησης κανονικοποιούνταν ανεξάρτητα, τότε θα μπορούσαν να προκύψουν διάφορα παράδοξα όπως το ακόλουθο: ανεξαρτήτως εάν η μέγιστη τιμή των δεδομένων αξιολόγησης ήταν X = 100 ή X = 100000, αυτή θα μετασχηματιζόταν στην τιμή 1 και θα ισοδυναμούσε με τη μετασχηματισμένη μέγιστη τιμή των δεδομένων εκπαίδευσης, ακόμη κι αν οι αρχικές τους τιμές είχαν διαφορά πολλών τάξεων μεγέθους.</font>

<font color='#486393'>Έχοντας τα κανονικοποιημένα δεδομένα, το επόμενο βήμα είναι η εκπαίδευση ενός μοντέλου γραμμικής παλινδρόμησης μέσω του αλγορίθμου Stochastic Gradient Descent (SGD). Γενικά, στόχος του προβλήματος της γραμμικής παλινδρόμησης είναι ο προσδιορισμός μιας ευθείας</font>

$$\hat{y} = w_0 + w_1x,$$

<font color='#486393'>προκειμένου το μέσο τετραγωνικό σφάλμα</font>

$$L\left(y,\hat{y}\right) = \frac{1}{N}\sum_{i=1}^{N}\left(y_i-\hat{y}_i\right)^2$$

<font color='#486393'>να ελαχιστοποιείται για το σύνολο των δεδομένων</font> $(x_i,y_i), i = 1, \dots, N$ <font color='#486393'>. Ο τρόπος με τον οποίο η ευθεία αυτή προσδιορίζεται μέσω του αλγορίθμου SGD είναι ο ακόλουθος: τα χαρακτηριστικά</font> $x_i$ <font color='#486393'>εισάγονται με τυχαίο τρόπο (στοχαστικά) διαδοχικά στο μοντέλο και για καθένα εξ αυτών υπολογίζεται η τιμή</font> $\hat{y}_i$<font color='#486393'>. Κατόπιν, τα βάρη </font> $w_0, w_1$ <font color='#486393'>ανανεώνονται σύμφωνα με τη σχέση</font>

$$w_j \rightarrow w_j - \eta \left.\frac{\partial L}{\partial w_j}\right|_{w_j},$$

<font color='#486393'>όπου </font>$\eta$ <font color='#486393'>είναι ο λεγόμενος ρυθμός εκμάθησης (learning rate). Η σχέση αυτή ισοδυναμεί με τις</font>

$$ w_0 \rightarrow w_0 - 2\eta \left(y_{i}-\hat{y}_i\right), \hspace{6mm} w_1 \rightarrow w_1 - 2\eta x_i \left(y_{i}-\hat{y}_i\right).$$

<font color='#486393'>Η διαδικασία εισαγωγής των δεδομένων και ανανέωσης των βαρών επαναλαμβάνεται μέχρι να εξαντληθούν όλα τα δείγματα. Λέμε τότε πως μια εποχή (epoch) εκπαίδευσης έχει ολοκληρωθεί. Ο τερματισμός του αλγορίθμου προκύπτει συνήθως με έναν από τους ακόλουθους τρόπους:</font>

- <font color='#486393'>με το πέρας ενός προκαθορισμένου πλήθους εποχών (αυτό που υλοποιείται στην προκείμενη περίπτωση, δεδομένης της εκφώνησης).</font>

- <font color='#486393'>όταν η διαφορά για το μέσο τετραγωνικό σφάλμα από το τέλος μια εποχής στο τέλος της επόμενης είναι μικρότερη από κάποιο κατώφλι, </font>$\epsilon$<font color='#486393'>.</font>

- <font color='#486393'>όταν το σφάλμα επικύρωσης φτάσει σε ελάχιστο.</font>

<font color='#486393'>Η τρίτη περίπτωση αποτελεί συχνά τον καλύτερο τρόπο τερματισμού του αλγορίθμου, όμως προϋποθέτει εκτός από τα δεδομένα εκπαίδευσης και αξιολόγησης ένα επιπλέον σύνολο: το σύνολο δεδομένων επικύρωσης. Αυτά, χρησιμοποιούνται μετά από κάθε κύκλο εκπαίδευσης ως ένα βήμα αξιολόγησης, προκειμένου να υπάρχει εποπτεία του κατά πόσο η γενίκευση του μοντέλου μετά από κάθε κύκλο εκπαίδευσης βελτιώνεται (δηλαδή το μοντέλο συνεχίζει να εκπαιδεύεται) ή χειροτερεύει (δηλαδή το μοντέλο ξεκινά να κάνει overfit). Ο λόγος για τον οποίο δεν υλοποιείται εδώ, είναι διότι ένα σύνολο επικύρωσης δεν προβλέπεται από την εκφώνηση της εργασίας.</font>

<font color='#486393'>Αφότου ολοκληρωθεί η εφαρμογή του SGD, το παραπάνω pipeline φροντίζει για την εκτύπωση της ευθείας στην οποία αντιστοιχούν τα υπολογισμένα βάρη, καθώς και του μέσου τετραγωνικού σφάλματος σε ό,τι αφορά τα δεδομένα αξιολόγησης. Κατόπιν, παρουσιάζονται δύο γραφικές παραστάσεις: μία όπου απεικονίζεται η εξέλιξη του μέσου τετραγωνικού σφάλματος με το πέρας κάθε εποχής, δίνοντας έτσι μια εικόνα για την πορεία εκπαίδευσης του μοντέλου, και μία όπου απεικονίζεται το σύνολο των δεδομένων αξιολόγησης μαζί με την υπολογισμένη ευθεία.</font>

<font color='#486393'>Τέλος, προκειμένου να υπάρχει εποπτεία του κατά πόσο ο αλγόριθμος SGD ήταν επιτυχής για την εκπαίδευση του μοντέλου γραμμικής παλινδρόμησης, υπολογίζεται ακριβώς (μέσω LMS) η ευθεία γραμμικής παλινδρόμησης που αντιστοιχεί στα δεδομένα και η εξίσωσή της τυπώνεται ως αναφορά.</font>

## Μέρος Β: Αποτελέσματα & Σχολιασμός

<font color='#486393'>Περνώντας στο δεύτερο μέρος, τα δεδομένα του αρχείου </font>`data1a.csv` <font color='#486393'>μεταφορτώνονται σε ένα pandas dataframe και περνάνε από το pipeline, για ρυθμό εκμάθησης ίσο με 0.03 και 10 εποχές εκπαίδευσης, δίνοντας τα ακόλουθα αποτελέσματα:</font>

```python
data = pd.read_csv('data1a.csv')
data = data[['likes','views']]

try:
    pipeline(data,0.2,0.03,10,True)
except ValueError:
    print("Εντοπίστηκε απειρισμός - ο αλγόριθμος τερματίστηκε.")
```

<font color='#486393'>Σε ό,τι αφορά το συντελεστή συσχέτισης ανάμεσα στα X, y, παρότι δεν υπάρχει ένα σαφές κατώφλι που να καθιστά τη συσχέτιση «πολύ» ή «λίγο» υψηλή, μπορεί κανείς να πει με ασφάλεια πως υπάρχει σημαντική θετική συσχέτιση ανάμεσα στις δύο μεταβλητές, ειδικά δεδομένου του πολύ υψηλού αριθμού δειγμάτων. Το γεγονός αυτό αποτυπώνεται και από το προκύπτον διάγραμμα διασποράς, όπου η θετική συσχέτιση μεταξύ αριθμού προβολών και αριθμού likes γίνεται ορατή. Παρ' όλα αυτά, δεν παύει να ισχύει πως υπάρχουν αρκετά σημεία τα οποία δημιουργούν μια απόκλιση από τη γραμμική συσχέτιση, γεγονός που γεννά την ανάγκη για επιπλέον ελέγχους σε ό,τι αφορά το είδος συσχέτισης των μεταβλητών.</font>

<font color='#486393'>Σχετικά με τον επιλεγόμενο αριθμό εποχών, το πρώτο από τα δύο τελικά γραφήματα υποδεικνύει πως η επιλογή ενός μικρού αριθμού είναι αιτιολογημένη, μιας και ο αλγόριθμος SGD οδηγεί σε σύγκλιση με ακρίβεια τάξης </font> $\mathcal{O}\left(10^{-4}\right)$ <font color='#486393'>μόλις μετά από 3-4 εποχές. Η τελική ευθεία παλινδρόμησης είναι η </font>$\hat{y} = 0.008 + 0.93x$ <font color='#486393'> και απεικονίζεται στο δεύτερο γράφημα, μαζί με τα αντίστοιχα σημεία αξιολόγησης, καθιστώντας εμφανή την προαναφερθείσα επιρροή ορισμένων σημείων, εμφανίζοντας όμως παράλληλα μια σχετικά καλή γραμμική προσαρμογή.</font>

<font color='#486393'>Συγκρίνοντας το αποτέλεσμα του SGD με αυτό που προκύπτει μέσω της μεθόδου ελαχίστων τετραγώνων (LMS), παρατηρείται μια απόκλιση ίση με 0.002 στην περίπτωση του σταθερού όρου και 0.003 στην περίπτωση του συντελεστή διεύθυνσης. Οι αποκλίσεις αυτές δεν είναι ιδιαίτερα υψηλές, υποδηλώνοντας έτσι την επιτυχία του SGD.</font>

<font color='#486393'>Ολοκληρώνοντας το σχολιασμό των αποτελεσμάτων για το πρώτο dataset, αξίζει να σταθούμε για λίγο στη σημασία του Min/Max Scaling (και γενικότερα της κανονικοποίησης). Εφαρμόζοντας το ίδιο pipeline, χωρίς όμως αυτή τη φορά την κανονικοποίηση των δεδομένων (αυτός είναι ο λόγος για τον οποίο η boolean μεταβλητή </font>`apply_minmax`<font color='#486393'> έχει εισαχθεί στο pipeline), προκύπτει το ακόλουθο:</font>

<font color='#486393'>Ο κώδικας αποτυγχάνει, εντοπίζοντας ένα ValueError κατά τον υπολογισμό του MSE για ένα ζεύγος δεδομένων. Όπως υποδεικνύει και το σχετικό RuntimeWarning, μιας και οι τιμές των δεδομένων ήταν πολύ υψηλές, ο αλγόριθμος οδηγήθηκε κάποια στιγμή σε απειρισμό, κάτι το οποίο φυσικά απεφεύχθη στην πρώτη περίπτωση, μέσω της κανονικοποίησης.</font>

Εδώ, δε χωρά αμφιβολία σε ό,τι αφορά τη συσχέτιση μεταξύ των δύο μεταβλητών και το γεγονός αυτό έρχεται να επαληθεύσει και το αντίστοιχο διάγραμμα διασποράς.</font>

<font color='#486393'>Επαναλαμβάνοντας την εκπαίδευση ενός μοντέλου γραμμικής παλινδρόμησης χρησιμοποιώντας τον SGD με ίδιο ρυθμό εκμάθησης και πλήθος εποχών, η προκύπτουσα ευθεία είναι η </font> $\hat{y} = 0.034 + 0.922x$<font color='#486393'> και απεικονίζεται στο 2ο γράφημα, μαζί με τα σημεία που αντιστοιχούν στο σύνολο αξιολόγησης. Σε ό,τι αφορά το μέσο σφάλμα ανά εποχή, γίνεται εμφανές πως η σύγκλιση πρακτικά επέρχεται ήδη με το πέρας της πρώτης εποχής, καθώς από εκείνη την εποχή κι έπειτα παρατηρούνται πολύ μικρές ταλαντώσεις του μέσου σφάλματος, οι οποίες είναι τάξης </font> $\mathcal{O}\left(10^{-6}\right)$<font color='#486393'>. Το γεγονός αυτό δεν είναι, φυσικά, τυχαίο - αντιθέτως οφείλεται στην υψηλή συσχέτιση των δύο μεταβλητών στην περίπτωση αυτή.</font>

<font color='#486393'>Σχετικά με την ευθεία που προκύπτει μέσω LMS, ο σταθερός όρος ταυτίζεται με αυτόν που προβλέπει ο SGD, ενώ στο συντελεστή διεύθυνσης υπάρχει μια απόκλιση ίση με 0.005. Παρ' όλα αυτά, ακριβώς λόγω της πολύ υψηλής συσχέτισης μεταξύ των δύο μεταβλητών, μειώνοντας το ρυθμό εκμάθησης (π.χ. στην τιμή 0.003) και αυξάνοντας ελαφρώς το πλήθος εποχών (π.χ. στην τιμή 25), παρατηρείται πλήρης ταύτιση των ευθειών που προκύπτουν με τις δύο διαφορετικές μεθόδους.</font>
